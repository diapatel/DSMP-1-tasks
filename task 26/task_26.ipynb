### **`Problem 5:`** Draw a strip plot and swarm plot between "gender" and "bloodpressure" w.r.t "smoker" (use as hue parameter). Also add title to the charts.

```python
# code here
insurance = pd.read_csv('./insurance_data - insurance_data (1).csv')
insurance.head()
```

```python
# stripplot
sns.stripplot(data=insurance, x='gender', y='bloodpressure', hue='smoker')
plt.title("Strip plot")
plt.show();
```

```python
sns.swarmplot(data=insurance, x='gender', y='bloodpressure')
plt.title("Swarm plot")
plt.show()
```

### **`Problem 6:`** Draw a Box-plot and a Violin plot of which x-axis represents the "region" and the y-axis represents the "bmi". Also add extra information of the column "diabetic".

```python
# code here
sns.boxplot(data=insurance, x='region', y='bmi', hue='diabetic')
plt.show()

# INTERPRETING THE BOXPLOT
# The median bmi is roughly the same whether the patient is diabetic or not in all regions
```

```python
sns.violinplot(data=insurance, x='region', y='bmi', hue='diabetic')
plt.show()
```

### **`Problem  7:`** Draw a bar plot and point plot of which x-axis represents the "gender" and y-axis represents "claim". Also add extra information about "smoker" column.

```python
# code here
sns.barplot(data=insurance, x='gender', y='claim', hue='smoker')
plt.show()
```

```python
sns.pointplot(data=insurance, x='gender', y='claim', hue='smoker')
plt.show()

# INTERPRETATION
# The claim amount for smokers and non smokers is significantly different but is roughly the same for both males and females
# if the person is a smoker, the claim is roughly around 8000-9000
# if the person is a non-smoker, the claim is around 31000-33000
```

### **`Problem 8:`** Draw a reg plot between "age" and "bmi" columns.

```python
# code here
sns.regplot(data=insurance, x='age', y='bmi')
plt.show()
```

### **`Problem 9:`** Draw a pair plot of the insurance data. Use "gender" as hue parameter.

```python
# code here
sns.pairplot(data=insurance, hue='gender')
plt.show()
```

### **`Problem 10:`** Draw a pair grid of the insurance data and use "diabetic" column as a hue parameter. Also, make the diagonal plots as box-plot, upper parts as scatter plot and the lower parts as kde plot.

```python
# code here
g = sns.PairGrid(data=insurance.drop(columns=['index', 'PatientID', 'gender']), hue='diabetic')
g.map_upper(sns.scatterplot)
g.map_diag(sns.boxplot)
g.map_lower(sns.kdeplot)
```

### **`Prolem 11:`** Draw a joint plot as scatter between "bloodpressure" and "bmi". Use "smoker" as hue parameter.
