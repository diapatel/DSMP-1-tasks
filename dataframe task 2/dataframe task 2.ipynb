###`Q-9:` IPL Point Table

Make a function `point_table` which take `season` as parameter and show points table in non-ascendng order of points and in ascending order of team name.

For winning - 2 Ponits;
For loosing - 0 Point
For not result both team gets 1 points.

Make dataframe which will have
`TeamName` `MatchesPlayed` `MatchesWon` `NoResult` `Points`
make `TeamName` as index.

```
season parametr should be one of these->
['2022', '2021', '2020/21', '2019', '2018', '2017', '2016', '2015', '2014', '2013', '2012', '2011', '2009/10', '2009', '2007/08']
```


Output of two Top 2 in season 2022
```
TeamName    MatchesPlayed	MatchesWon	NoResult	Points

Gujarat Titans	    16	   12	       0	     24
Rajasthan Royals	  17	   10	       0	     20

```

````python
ipl.columns
````

````python
def points_table(season):

  """This function takes in a year as teh season value and returns a dataframe with the following attributes:
  1. team name
  2. number of matches played in that season
  3. number of matches won
  4. number of matches wiht no results as the outcome
  5. points: where for winning a match, 2 points are rewarded per match and for no result 1 point is rewarded."""

  if type(season) == int:
    season = str(season)
  season_df = ipl[ipl['Season'] == season]
  teams_list = ipl['Team1'].unique()
  match_count = []
  win_count = []
  no_result_count = []
  points_list = []

  for team in teams_list:
    temp_df = season_df[(season_df['Team1'] == team) | (season_df['Team2'] == team)]
    matches_played = temp_df['ID'].count()
    matches_won = temp_df[temp_df['WinningTeam'] == 'Gujarat Titans']['ID'].count()
    no_results = temp_df[temp_df['WonBy'] == 'NoResults']['ID'].count()
    points = (matches_won*2) + (1*no_results)

    match_count.append(matches_played)
    win_count.append(matches_won)
    no_result_count.append(no_results)
    points_list.append(points)

  # create and populate the dataframe
  result = pd.DataFrame()
  result['team'] = teams_list
  result['matches_played'] = match_count
  result['matches_won'] = win_count
  result['no_result'] = no_result_count
  result['points'] = points_list

  return result
````
